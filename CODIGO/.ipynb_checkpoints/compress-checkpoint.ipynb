```python
import urllib.request as urlib
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.fftpack import dct, idct
import cv2
from math import floor
```

```python
def transform2D(img,function):
    return np.transpose(function(np.transpose(function(img,norm = "ortho")),norm = "ortho"))
    
```

```python
def segmentation(img,size,function=transform2D,function2 =dct):
    copy = np.zeros(img.shape)
    k =1;
    #print(copy.shape[0]//size)
    for i in range(copy.shape[0]//size):
        for j in range(copy.shape[1]//size):
            copy[i*size:(i+1)*size,j*size:(j+1)*size] = function(img[i*size:(i+1)*size,j*size:(j+1)*size],function2)
            #k+=1;
            #print(i*size,(i+1)*size,j*size,(j+1)*size)
            #print(function(img[i*size:(i+1)*size,j*size:(j+1)*size],function2))
    return copy
```

```python
def thresholding(img,nthV):
    imgCopy = img.copy()
    sortedCts = np.sort(abs(imgCopy.ravel()))
    threshold = sortedCts[-nthV]
    imgCopy[abs(imgCopy) < threshold] = 0;
    print(threshold)
    return imgCopy
    
    
```

```python
def percent(img,value):
    return floor((img.shape[0]*img.shape[1])*(value/100))
```

```python
img1 = segmentation(img,8)
img2 = thresholding(img1,percent(img,10))
np.savetxt('img2.out', img2)

img3 = segmentation(img2,8,function2 = idct)
plt.imshow(img3)
cv2.imwrite('compresed.jpg',img3)
np.savetxt('img3.out', img3)
print(img3)
```
